```python
#K-40 Activity 
fgcps = []
netcps = []
fgunc = []
unc_net = []
Activities = []
Unc_A = []
eff = 0.01
Br = 0.1
#sample 4
n = Spec[3]
cps = n[0][0]/t4
unc = n[1][0]/t4
fgcps.append(cps)
fgunc.append(unc)
#sample 6
n = Spec[5]
cps = n[0][0]/t6
unc = n[1][0]/t6
fgcps.append(cps)
fgunc.append(unc)
#sample 7
n = Spec[6]
cps = n[0][0]/t7
unc = n[1][0]/t7
fgcps.append(cps)
fgunc.append(unc)
#sample 9
n = Spec[8]
cps = n[0][0]/t9
unc = n[1][0]/t9
fgcps.append(cps)
fgunc.append(unc)
#sample 10
n = Spec[9]
cps = n[0][0]/t10
unc = n[1][0]/t10
fgcps.append(cps)
fgunc.append(unc)

print(fgcps)
print(fgunc)
```

```python
import numpy as np
#background 1 
b = Bg[0]
for i in range(0,5):
    net_cps = fgcps[i] - b[0][0]/tbg1
    uncnet = np.sqrt((fgunc[i])**2 + (b[1][0]/tbg1)**2)
    netcps.append(net_cps)
    unc_net.append(uncnet)
    Activity = net_cps/(eff*Br)
    unc_A = uncnet/(eff*Br)
    Activities.append(Activity)
    Unc_A.append(unc_A)

print(netcps)
print(unc_net)
print(Activities)
print(Unc_A)
```
